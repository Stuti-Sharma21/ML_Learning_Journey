```python
# dataset
hours = [1, 2, 3, 4]
marks = [40, 50, 60, 70]

# starting values
m = 0
b = 0

learning_rate = 0.01

# training loop
for epoch in range(100):

    total_error = 0

    for i in range(len(hours)):

        x = hours[i]
        actual_y = marks[i]

        # prediction
        predicted_y = m * x + b

        # error
        error = actual_y - predicted_y

        # update parameters
        m = m + learning_rate * error * x
        b = b + learning_rate * error

        total_error += error ** 2

    print("Epoch:", epoch)
    print("m:", m)
    print("b:", b)
    print("Error:", total_error)
    print()
```

```python
# dataset
hours = [1, 2, 3, 4]
marks = [40, 50, 60, 70]

# starting values
m = 0
b = 0

learning_rate = 0.1

# training loop
for epoch in range(100):

    total_error = 0

    for i in range(len(hours)):

        x = hours[i]
        actual_y = marks[i]

        # prediction
        predicted_y = m * x + b

        # error
        error = actual_y - predicted_y

        # update parameters
        m = m + learning_rate * error * x
        b = b + learning_rate * error

        total_error += error ** 2

    print("Epoch:", epoch)
    print("m:", m)
    print("b:", b)
    print("Error:", total_error)
    print()
```

```python
import matplotlib.pyplot as plt

# dataset
hours = [1, 2, 3, 4]
marks = [40, 50, 60, 70]

# learned parameters
m = 10
b = 29

# predictions
predictions = []

for x in hours:
    y = m * x + b
    predictions.append(y)

# scatter plot (actual points)
plt.scatter(hours, marks)

# regression line
plt.plot(hours, predictions)

# labels
plt.xlabel("Hours Studied")
plt.ylabel("Marks")

# title
plt.title("Linear Regression")

# show graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# dataset
hours = np.array([1, 2, 3, 4])
marks = np.array([40, 50, 60, 70])

# parameters
m = 10
b = 30

# predictions
predictions = m * hours + b

# graph
plt.scatter(hours, marks)
plt.plot(hours, predictions)

plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Linear Regression with NumPy")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

# dataset
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)

marks = np.array([40, 50, 60, 70, 80])

# model
model = LinearRegression()

# training
model.fit(hours, marks)

# predictions
predictions = model.predict(hours)

# learned parameters
print("Slope:", model.coef_)
print("Intercept:", model.intercept_)

# graph
plt.scatter(hours, marks)

plt.plot(hours, predictions)

plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Linear Regression")

plt.show()
```
